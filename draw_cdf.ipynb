```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def plot_cdf_side(data1, data2, title):
    plt.subplot(1,2,1)
    plt.title(title)
    values, base = np.histogram(data1, bins=100)
    cumulative = np.cumsum(values)
    plt.plot(base[:-1], cumulative, c='blue')
    
    plt.subplot(1,2,2)
    values, base = np.histogram(data2, bins=100)
    cumulative = np.cumsum(values)
    plt.plot(base[:-1], cumulative, c='blue')
    plt.tight_layout()
    plt.show()

def plot_cdf_single(data, title):
    values, base = np.histogram(data, bins=100)
    cumulative = np.cumsum(values)
    plt.plot(base[:-1], cumulative, c='blue')
    plt.title(title)
    plt.show()
```

```python
# import data from file
reader = csv.reader(open('var-master-original-new/output-small.csv', 'r'), delimiter=' ')
master_data = np.array(list(reader)).astype('int')
reader = csv.reader(open('var-development-original-new/output-small.csv', 'r'), delimiter=' ')
development_data = np.array(list(reader)).astype('int')

plot_cdf_side(master_data[:,0], development_data[:,0], 'txnAttempts')
plot_cdf_side(master_data[:,1], development_data[:,1], 'Backoff Time')
plot_cdf_side(master_data[:,2], development_data[:,2], 'Backoff Count')
```

```python
def count_elements(data):
    hist = {}
    for i in range(np.shape(data)[0]):
        j = data[i, 0]
        hist[j] = hist.get(j, 0) + 1
    return hist
```
